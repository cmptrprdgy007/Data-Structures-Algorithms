```python
class Node1:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DLL:
    def __init__(self):
        self.head = None

    def forward_traversal(self):
        print()
        if self.head is None:
            print('Empty linked list!!!')
        else:
            a = self.head
            while a is not None:
                print(a.data, end=' ')
                a = a.next
        print()  # To ensure the output ends with a newline

    def backward_traversal(self):
        print()
        if self.head is None:
            print('Empty linked list!!!')
        else:
            a = self.head
            while a.next is not None:
                a = a.next
            while a is not None:
                print(a.data, end=' ')
                a = a.prev
        print()  # To ensure the output ends with a newline

    def middle(self):
        if self.head is None:
            return None
        
        slow = self.head
        fast = self.head

        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

        return slow.data if slow else None

    def cycle_det(self):
        if self.head is None:
            return 'No nodes in the list'
        
        slow = self.head
        fast = self.head

        while fast is not None and fast.next is not None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:  # Cycle detected
                return 'Cycle detected'
        
        return 'No cycles detected'

# Example usage:
dll = DLL()
n1 = Node1(4)
dll.head = n1
n2 = Node1(90)
n2.prev = n1
n1.next = n2
n3 = Node1(6)
n3.prev = n2
n2.next = n3
n3.next = n1  # Creating a cycle

# Testing the functions
dll.forward_traversal()
print("Middle node:", dll.middle())
print(dll.cycle_det())  # Should output 'Cycle detected'
```
