```python
def fact(num):
    if num == 0:
        return 1
    return num * fact(num-1)
fact(5)
```

```python
# Power exponential using recursion

def power(base ,exp):
    if exp == 0:
        return 1
    return base * power(base,exp -1)
power(4,6)
```

```python
# print something n times

def print_(n):
    print(n)
    return print_(n)
#print_(9)
```

```python
# print  1 to n using recursion

def print_1(n):
    if n <1 :
        return
    print(n)
    print_1(n-1)
   
print_1(6)
```
