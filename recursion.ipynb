```python
def fact(num):
    if num == 0:
        return 1
    return num * fact(num-1)
fact(5)
```

```python
# Power exponential using recursion

def power(base ,exp):
    if exp == 0:
        return 1
    return base * power(base,exp -1)
power(4,6)
```

```python
# print something n times

def print_(n):
    print(n)
    return print_(n)
#print_(9)
```

```python
# print  n to 1 using recursion

def print_1(n):
    if n <1 :
        return
    print(n)
    print_1(n-1)
   
print_1(6)
```

```python
def print_1_to_n(n):
    if n < 1:
        return
    print_1_to_n(n - 1)  # First, call the function recursively
    print(n)
print_1_to_n(5)
```

```python
# sum of first n number - normal way

def sum_n(n):
    sum = (n *(n+1)) / 2
    return sum
sum_n(6)
```

```python
# Recursive way to print sum of 1st n numbers

def sum_n1(n):
    if n == 0:
        return 0
    return n + sum_n1(n-1)
sum_n1(6)

```

```python
# reverse an array 

def rev(arr,l,r):
    if l >= r: # stops at middle index
        return 
    arr[l],arr[r] = arr[r],arr[l]
    r = len(arr) - 1 # swap
    return rev(arr,l+1,r-1)
arr = [1,2,3,4]
rev(arr,0,len(arr) - 1)
print(arr)
    
```

```python
arr = [1, 2, 3, 4, 5]
rev(arr, 0, len(arr) - 1)
print(arr)  # Output: [5, 4, 3, 2, 1]
```

```python
def palin(s, i=0):
    # Base case: if we've checked all characters
    if i >= len(s) // 2:
        return True # it reached at middle element then we stops it
    # Compare characters from the start and end
    if s[i] != s[len(s) - 1 - i]:
        return False
    # Recursive call for the next pair of characters
    return palin(s, i + 1)
palin('madam')
palin('akshay')
```

```python
def fib(n):
    if n <= 0:
        return []  # Return an empty list for non-positive input
    elif n == 1:
        return [0]  # Return only the first Fibonacci number
    elif n == 2:
        return [0, 1]  # Return the first two Fibonacci numbers

    # Initialize the sequence with the first two Fibonacci numbers
    seq = [0, 1]
    for i in range(2, n+1):
        next_fib = seq[i-1] + seq[i-2]  # Calculate the next Fibonacci number
        seq.append(next_fib)  # Add it to the sequence

    return seq[-1]

# Example Usage:
print(fib(7))  # Output: [0, 1, 1, 2, 3, 5, 8, 13, 21]
```

```python
def fib_(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:

        last = fib_(n-2)
        first = fib_(n-1)
        return last+first
fib_(7)
```
